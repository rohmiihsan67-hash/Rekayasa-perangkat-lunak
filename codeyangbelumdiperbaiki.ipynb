```python
from dataclasses import dataclass

# MODEL DATA
@dataclass
class Order:
    customer_name: str
    total_price: float
    status: str = "open"


# KODE BURUK (MELANGGAR CLEAN CODE)
class OrderManager:
    def process_checkout(self, order: Order, payment_method: str):
        print(f"Memulai checkout untuk {order.customer_name}...")

        # LOGIKA PEMBAYARAN (Melanggar OCP & DIP)
        if payment_method == "credit_card":
            print("Processing Credit Card...")
        elif payment_method == "bank_transfer":
            print("Processing Bank Transfer...")
        elif payment_method == "qris":
            print("Processing QRIS...")
        else:
            print("Metode pembayaran tidak valid.")
            return False

        # LOGIKA NOTIFIKASI (Melanggar SRP)
        print(f"Mengirim notifikasi ke {order.customer_name}...")

        # UPDATE STATUS
        order.status = "paid"

        print("Checkout selesai.\n")
        return True


# PROGRAM UTAMA
def main():
    manager = OrderManager()

    # Skenario 1
    order1 = Order("Andi", 500000)
    manager.process_checkout(order1, "credit_card")

    # Skenario 2
    order2 = Order("Budi", 100000)
    manager.process_checkout(order2, "qris")

    # Skenario 3
    order3 = Order("Siti", 250000)
    manager.process_checkout(order3, "bank_transfer")


if __name__ == "__main__":
    main()
```
